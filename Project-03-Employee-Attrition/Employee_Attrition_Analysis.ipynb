```python
# Hypothesis Testing
# Determine if the observed differences are statistically significant (real) or just due to random chance.
```

```python
# T-Test (Salary vs Attrition)
# Do employees who leave have lower salary?
# Null Hypothesis (H0):
# Salary of employees who leave and stay is same

# Alternative Hypothesis (H1):
# Salary of employees who leave and stay is different
```

```python
from scipy import stats
```

```python
employee_stay = df[df['Attrition'] == 'No']['MonthlyIncome']
employee_leave = df[df['Attrition'] == 'Yes']['MonthlyIncome']
```

```python
t_stat,p_value = stats.ttest_ind(employee_stay,employee_leave)
```

```python
print(t_stat)
print(p_value)
```

```python
# since p_value < 0.05 we reject null hypothesis salary affects attrition
```

```python
# Chi-Square Test
# Does overtime affect attrition?
#Null Hypothesis (H0):
#Overtime and Attrition are independent(No relationship)

#Alternative Hypothesis (H1):
#Overtime and Attrition are related 
```

```python
cont_table = pd.crosstab(df['Attrition'],df['OverTime'])
```

```python
chi2,p_value,dof,eop = stats.chi2_contingency(cont_table)
```

```python
print(eop)
print(cont_table)
```

```python
chi2
```

```python
print(p_value)
```

```python
#since p_value < 0.05 we reject null hoypthiesis hence overtime and attrition are related 
# Employees who work overtime are more likely to leave the company.
```

```python
# ANOVA Test
# Does salary differ across departments?
# H0: All departments have same salary mean
# H1: At least one department has different salary mean
```

```python
df['Department'].value_counts()
```

```python
groups = [group['MonthlyIncome'].values for name,group in df.groupby('Department')]
```

```python
f_stat,p_value = stats.f_oneway(*groups)
```

```python
print(f_stat)
print(p_value)
```

```python
# we reject null hypothesis 
#This suggests that department plays an important role in determining employee salary.
```

```python
df['Attrition'] = df['Attrition'].map({'Yes':1,'No':0})
```

```python
df.head(2)
```

```python
df.dtypes
```

```python
df.drop(columns = ['EmployeeCount','EmployeeNumber','Over18','StandardHours'],inplace = True)
```

```python
df = pd.get_dummies(df,drop_first = True)
```

```python
df.head(2)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X = df.drop(['Attrition'],axis = 1)
y = df['Attrition']
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 7)
```

```python
lr = LogisticRegression(max_iter = 1000)
```

```python
lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,roc_curve,auc
```

```python
y_pred = lr.predict(x_test)
```

```python
accuracy = accuracy_score(y_test,y_pred)
```

```python
print('accuracy:',accuracy)
print('ClassificationReport:',classification_report(y_test,y_pred))
print('ConfusionMatrix:',confusion_matrix(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(n_estimators = 100,random_state = 7)
```

```python
rf_model.fit(x_train,y_train)
```

```python
pred1 = rf_model.predict(x_test)
```

```python
print('accuracy:',accuracy_score(y_test,pred1))
print('ClassificationReports:',classification_report(y_test,pred1))
print('ConfusionMatrix:',confusion_matrix(y_test,pred1))  
```

```python
#Recall = How many actual Yes caught

#Precision = How many predicted Yes correct
```

```python
rf_model = RandomForestClassifier(n_estimators = 200,class_weight = 'balanced',random_state = 7)
```

```python
rf_model.fit(x_train,y_train)
```

```python
pred1 = rf_model.predict(x_test)
```

```python
print('Accuracy:',accuracy_score(y_test,pred1))
print('Classification:',classification_report(y_test,pred1))
print('ConfusionMatrix:',confusion_matrix(y_test,pred1))
```

```python
'''## Model Comparison and Final Selection

Multiple models were evaluated including Logistic Regression and Random Forest.

Although Random Forest achieved slightly higher accuracy, Logistic Regression achieved better recall for predicting employee attrition.

Since the business objective is to identify employees likely to leave, recall is the most important metric.

Therefore, Logistic Regression is selected as the final model.'''
```
